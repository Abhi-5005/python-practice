```python
#  Lambda Functions
"""
Its small anonymous function
lambda function is one time use function,
single line code
temoparary use code
single expression
just in time
lambda arguments : expression
"""
```

```python
(lambda x,y : x+y)(10,20) 
```

```python
(lambda x,y: 'x' if x>y else 'y')(50,30)
```

```python
x = lambda x:x+20
```

```python
print(x(20))
```
