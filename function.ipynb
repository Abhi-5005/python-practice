```python
# Write a function to return the factorial of a number

def factorial(num):
    fac = 1
    for i in range(num,1,-1):
        fac = fac*i
    return fac 
```

```python
#  Write a function to count vowels in a string.
def vowels():
    count = 0
    string = "aeiholooo"
    for i in string:
        if i== 'a' or i== 'e' or i== 'i' or i== 'o' or i== 'u':
            count+=1
        else:
            continue
    return count
```

```python
# Write a function to check if a string is a palindrome.

def palindrome():
    string = "abhi"
    rev= ""
    for i in range(-1,-len(string)-1,-1):
        rev = rev + string[i]
    return rev
        
```

```python
# Write a function that checks if a number is an Armstrong number.

# 153 = 1*1*1 + 5*5*5 + 3*3*3 = 153

def arm(num):
    temp = num
    digits = len(str(num))
    sum1 = 0
    for i in range(digits):
        digit = temp%10
        sum1 += digit**digits
        temp = temp//10
    if sum1 == num:
        print("armstrong")
    else:
        print("not armstrong")
```

```python
# Write a function to check if a number is perfect.

# 6  = 1+2+3 = 6

def perfect(num):
    sum1 = 0
    for i in range(1,num):
        if num % i == 0:
            sum1 += i
    if num == sum1:
        print("perfect")
    else:
        print("Not perfect")
```

```python
perfect(7)
```

```python
import time

def timer(fun):
    def inner(*args, **kwargs):
        start = time.time()
        result = fun(*args, **kwargs)
        end = time.time()
        print(f"Execution time: {end - start}")
        return result
    return inner

@timer
def work():
    for i in range(1000000):
       pass

work()
```

```python
# 1 üß© Write a decorator that prints ‚ÄúStart‚Äù before and ‚ÄúEnd‚Äù after a function runs.
from functools import wraps
def deco(fun):
    @wraps(fun)
    def wrapper():
        print("start")
        fun()
        print("after")
    return wrapper

@deco
def greet():
    print("run")


print(greet.__name__)
```

```python
# 2 üß© Decorate a function that adds two numbers. The decorator should print "Running addition..." before it runs.

def deco(fun):
    def wrapper(*x,**y):
        print("Running addition...")
        fun(*x,**y)
    return wrapper


@deco
def add(a,b):
    print(a+b)


add(10,20)
```

```python
# 3 üß© Create a decorator that doubles the return value of the decorated function.

def deco(fun):
    def inner():
        r = 2*fun()
        return r
    return inner

@deco
def double():
    return 10

double()
```

```python
# 4 üß© Write a decorator that checks if the argument passed is positive. If not, print "Invalid number" instead of running the function.

def deco(fun):
    def inner(a,b):
        if (a < 0 or b<0):
            print("Invalid")
        else:
            fun(a,b)
    return inner

@deco
def check(x,y):
    print(x+y)


check(10,20)
```
